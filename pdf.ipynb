```python
count,bins=np.histogram(iris['PetalLengthCm'])
print(count)
print(bins)

pdf=count/sum(count)
print(pdf)
#plt.plot(bins[:-1],pdf)
cdf=np.cumsum(pdf)
print(cdf)
plt.plot(bins[:-1],pdf)
plt.plot(bins[:-1],cdf)
```

```python
count,bins=np.histogram(iris[iris['Species']=='Iris-setosa']['PetalLengthCm'])
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bins[:-1],pdf)
plt.plot(bins[:-1],cdf)
count1,bins1=np.histogram(iris[iris['Species']=='Iris-versicolor']['PetalLengthCm'])
pdf1=count/sum(count)
cdf1=np.cumsum(pdf)
plt.plot(bins1[:-1],pdf1)
plt.plot(bins1[:-1],cdf1)
count2,bins2=np.histogram(iris[iris['Species']=='Iris-virginica']['PetalLengthCm'])
pdf2=count/sum(count)
cdf2=np.cumsum(pdf)
plt.plot(bins2[:-1],pdf2)
plt.plot(bins2[:-1],cdf2)
```
